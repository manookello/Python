```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.kilimall.co.ke/search-result?keyword=hp+laptops&source=search|enterSearch|hp+laptops'
page = requests.get(url)

headers={"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:129.0) Gecko/20100101 Firefox/129.0","Accept-Encoding":"gzip, deflate, br, zstd","Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/png,image/svg+xml,*/*;q=0.8","Upgrade-Insecure-Requests":"1"}

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
soup2 = BeautifulSoup(soup.prettify(), "html.parser")
```

```python
body=soup2.find('body')
print(body)
```

```python
name=soup2.find_all('p',class_="product-title")
print(name)
```

```python
names=[name.text.strip() for name in name]
print(names)
```

```python
price=soup2.find_all('div',class_="product-price")
print(price)
```

```python
prices=[price.text.strip() for price in price]
print(prices)
```

```python
dict={"name":names,"price":prices}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dict)
```
