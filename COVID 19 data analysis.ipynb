```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r"/home/okello/Desktop/Analyst Builder/CovidDeaths.xlsx")
df
```

```python
pd.set_option("display.float_format", lambda x: "%.2f" % x)
```

```python
df.describe()
```

```python
#looking at total_deaths vs total_cases
continent_summary=df.groupby("continent").agg({"total_cases":"sum","total_deaths":"sum"}).reset_index()
continent_summary["Death_Rate(%)"]=(continent_summary["total_deaths"] / continent_summary["total_cases"])*100
print(continent_summary)
```

```python
#looking at total_cases vs population
continent_summary_population=df.groupby("continent").agg({"total_cases":"sum","population":"sum"}).reset_index()
continent_summary_population["Death_Rate_population(%)"]=(continent_summary_population["total_cases"] / continent_summary_population["population"])*100
print(continent_summary_population)
```

```python
continent_deaths=df.groupby("continent")["total_cases"].sum().reset_index()
```

```python
continent_deaths=continent_deaths.sort_values(by="total_cases", ascending=False)
```

```python
plt.figure(figsize=(10,6))
plt.bar(continent_deaths["continent"], continent_deaths["total_cases"], color="skyblue")
plt.xlabel("Continent")
plt.ylabel("Total Deaths")
plt.title("Total COVID-19 Deaths perContinent")
plt.xticks(rotation=45)
plt.show()
```

```python
total_deaths=continent_deaths["total_cases"].sum()
continent_deaths["Percentage"]=(continent_deaths["total_cases"]/total_deaths)*100
```

```python
plt.figure(figsize=(8,8))
plt.pie(continent_deaths["Percentage"], labels=continent_deaths["continent"],autopct="%1.1f%%", startangle=140, colors=plt.cm.Paired.colors)
plt.title("Percentage of Total COVID-19 Deaths per Continent")
plt.show()
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df=df.sort_values("date")
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["date"],df["total_cases"],label="Total Cases")
plt.xlabel("Date")
plt.ylabel("Total_Cases")
plt.title("Total COVID 19 Cases Over Time")
plt.legend()
plt.show()
```

```python
df["7_day_avg_new_cases"]=df["new_cases"].rolling(window=7).mean()
plt.figure(figsize=(10,5))
plt.plot(df["date"],df["new_cases"], label="New Cases", color = "red", alpha=0.5)
plt.plot(df["date"], df["7_day_avg_new_cases"], label="7-Day Moving Averages", color="blue")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.title("7-Day Moving Average of New COVID 19 Cases")
plt.legend()
plt.show()
```

```python
df["daily_growth_rate"]=df["new_cases"].pct_change()*100
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["date"],df["daily_growth_rate"], label="Daily Growth Rate", color="green")
plt.xlabel("Date")
plt.ylabel("Daily Growth Rate (%)")
plt.title("Daily Growth Rate of New COVID 19 Cases")
plt.legend()
plt.show()
```
