```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://jiji.co.ke/cars?query=bmw%20%20x5'
page = requests.get(url)

headers={"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:129.0) Gecko/20100101 Firefox/129.0","Accept-Encoding":"gzip, deflate, br, zstd","Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/png,image/svg+xml,*/*;q=0.8","Upgrade-Insecure-Requests":"1"}

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
soup2 = BeautifulSoup(soup.prettify(), "html.parser")
```

```python
car_name=soup2.find_all('div',class_='b-advert-title-inner qa-advert-title b-advert-title-inner--div')
car_name
```

```python
car_names=[car_name.text.strip() for car_name in car_name]
print(car_names)
```

```python
price=soup2.find_all('div',class_='qa-advert-price')
print(price)
```

```python
prices=[price.text.strip() for price in price]
print(prices)
```
