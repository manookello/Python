```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.jumia.co.ke/elitebook-820-g1-intel-core-i5-6th-gen-8gb-ram-256gb-ssd-12.5-inches-hp-mpg604961.html"
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
title = soup.find('title')
```

```python
print(title)
```

```python
hp_title=title.text
```

```python
print(hp_title)
```

```python
title = soup.find('div', class_='-hr -mtxs -pvs')
```

```python
print(title)
```

```python
price=title.find('span').text
```

```python
print(price)
```
